```
from pyspark.sql import SparkSession
from pyspark.sql.functions import upper,to_date,col,year,month,dayofmonth
```

```
spark
```

```
spark = SparkSession.builder.getOrCreate()
```

```
bucket = 'datalake-jmsp-dmc9'
ruta_bronze = f"gs://{bucket}/bronze/sap/transacciones/"
ruta_silver = f"gs://{bucket}/silver/sap/transacciones/"

df = spark.read.format("parquet").option("header","true").load(ruta_bronze)
display(df)
```

```
df.printSchema()
```

```
df_transaccion_procesado = df.withColumn("ID_PERSONA", col("ID_PERSONA").cast("Integer")) \
.withColumn("ID_EMPRESA", col("ID_EMPRESA").cast("Integer")) \
.withColumn("FECHA", to_date(col("FECHA"), "yyyy-MM-dd"))\
.withColumn("YEAR",year(col("FECHA")))\
.withColumn("MONTH",month(col("FECHA")))\
.withColumn("DAY",dayofmonth(col("FECHA")))\

```

```
df_transaccion_procesado.write.mode("overwrite").format("delta").partitionBy("YEAR","MONTH","DAY").save(ruta_silver)

```
